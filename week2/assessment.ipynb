```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('/Users/anusha/Desktop/Projects/CV/Computer-Vision-with-Python/DATA/dog_backpack.jpg')
```

```python
plt.imshow(img)
```

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
flipped_img = cv2.flip(fix_img,0)
plt.imshow(flipped_img)
```

```python
cv2.rectangle(fix_img,pt1=(200,400),pt2=(600,700),color=(255,0,0),thickness=8)
plt.imshow(fix_img)
```

```python
vertices = np.array([[400,400],[250,700],[600,700]],dtype=np.int32)
pts = vertices.reshape(-1,1,2)
cv2.polylines(fix_img,[pts],isClosed=True,color=(0,0,255),thickness=5)
plt.imshow(fix_img)
```

```python
vertices = np.array([[400,400],[250,700],[600,700]],dtype=np.int32)
pts = vertices.reshape(-1,1,2)
cv2.fillPoly(fix_img,[pts],color=(0,0,255))
plt.imshow(fix_img)
```

```python
def draw_circle(event,x,y,flags,param):
        
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),100,(255,0,0),5)

img = cv2.imread('/Users/anusha/Desktop/Projects/CV/Computer-Vision-with-Python/DATA/dog_backpack.jpg')

cv2.namedWindow(winname='DogBackpack')

cv2.setMouseCallback('DogBackpack',draw_circle)



while True:
    
    cv2.imshow('DogBackpack',img)
    
    if cv2.waitKey(20) & 0xFF ==27:
        break
        
cv2.destroyAllWindows()
```
