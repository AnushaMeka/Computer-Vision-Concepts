# Smoothing Images
The Median filter is a non-linear filter that is most commonly used as a simple way to reduce noise in an image. It's claim to fame (over Gaussian for noise reduction) is that it removes noise while keeping edges relatively sharp.

```python
from skimage.filters import median

image = io.imread('/Users/anusha/Desktop/Projects/CV/Computer-Vision-with-Python/DATA/Denis_Mukwege.jpg', as_gray=True)
image_median = median(image)

plt.subplot(121), plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(122), plt.imshow(image_median, cmap='gray')
plt.title('Smooth Image')

plt.show()
```

# Blurring Images 
A Gaussian filter is a linear filter. It's usually used to blur the image or to reduce noise. If you use two of them and subtract, you can use them for "unsharp masking" (edge detection). The Gaussian filter alone will blur edges and reduce contrast.

```python
from skimage.filters import gaussian
image = io.imread('/Users/anusha/Desktop/Projects/CV/Computer-Vision-with-Python/DATA/sudoku.jpg', as_gray=True)
blurred_img = gaussian(image, 2, multichannel=True, mode='reflect')
plt.subplot(121), plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(122),plt.imshow(blurred_img, cmap='gray')
plt.title('Blurred Image')
```

```python
import cv2
cv2_response = cv2.GaussianBlur(image, (5,5), 2)
plt.subplot(121), plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(122),plt.imshow(cv2_response, cmap='gray')
plt.title('cv2 Blurred Image')
#plt.imshow(cv2_response, cmap='gray')
```

```python
# https://scikit-image.org/docs/dev/auto_examples/applications/plot_rank_filters.html#sphx-glr-auto-examples-applications-plot-rank-filters-py
from skimage.util import img_as_ubyte
from skimage import data
from skimage.exposure import histogram

noisy_image = img_as_ubyte(data.camera())
hist, hist_centers = histogram(noisy_image)

fig, ax = plt.subplots(ncols=2, figsize=(10, 5))

ax[0].imshow(noisy_image, cmap=plt.cm.gray)
ax[0].axis('off')

ax[1].plot(hist_centers, hist, lw=2)
ax[1].set_title('Histogram of grey values')

plt.tight_layout()
```
